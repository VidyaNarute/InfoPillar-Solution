# o Canny edge detection:

#### The Canny edge detection algorithm is composed of 5 steps:  
Noise reduction, Gradient calculation, Non-maximum suppression,Double threshold, Edge Tracking by Hysteresis

```python
import cv2
import matplotlib.pyplot as plt
# Open the image
img = cv2.imread('Abdul Kalam Sir.jpg')
# Apply Canny
edges = cv2.Canny(img, 50, 100, 3, L2gradient=True)
plt.figure()

plt.figure(figsize=(8.5,8.5))
plt.subplot(1,2,1),plt.imshow(img),plt.title("GRAY")
plt.imsave('Kalam-canny.png', edges, cmap='gray', format='png')
plt.imshow(edges, cmap='gray')
plt.subplot(1,2,2),plt.imshow(img),plt.title("BrBG")
plt.imsave('Kalam-canny.png', edges, cmap='BrBG', format='png')
plt.imshow(edges, cmap='BrBG')
plt.show()
plt.show()
```

```python
img = cv2.imread('Thumbs.jpg')
plt.imshow(img)
#Converting it to RGB and then grayscale:

# convert to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### We gonna need to create a binary image, which means each pixel of the image is either black or white. This is a necessity in OpenCV, finding contours is like finding a white object from a black background, objects to be found should be white and the background should be black.

```python
# create a binary thresholded image
_, binary = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)
# show it
plt.imshow(binary, cmap="gray")
plt.show()
```

### Now the above code creates the binary image by disabling (setting to 0) pixels that have a value of less than 225 and turning on (setting to 255) the pixels that has a value of more than 225, here is the output image:

```python
#Now, this is easy for OpenCV to detect contours:

# find the contours from the thresholded image
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# draw all contours
img = cv2.drawContours(img, contours, -1, (0, 255, 0), 2)

#Now this code finds contours within the binary image and draws them with a thick green line to the image, let's show it:
# show the image with the drawn contours
plt.imshow(img)
plt.show()
```
